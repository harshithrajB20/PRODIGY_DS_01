```python
#TASK NO:1
import matplotlib.pyplot as plt
from collections import Counter

def get_entity_data():
    try:
        n = int(input("How many people do you want to enter data for? "))
        if n <= 0:
            raise ValueError
    except ValueError:
        print(" Please enter a valid positive integer.")
        return get_entity_data()

    age_data = []
    gender_data = []

    print("\nEnter details for each person:")
    for i in range(1, n + 1):
        try:
            age = int(input(f"Person {i} - Age: "))
            gender = input(f"Person {i} - Gender (Male/Female/Other): ").strip().capitalize()
            age_data.append(age)
            gender_data.append(gender)
        except ValueError:
            print(" Invalid input. Try again.")
            return get_entity_data()

    return age_data, gender_data

def plot_age_histogram(age_data):
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.hist(age_data, bins=10, color='skyblue', edgecolor='black')
    plt.title("Age Distribution")
    plt.xlabel("Age")
    plt.ylabel("Frequency")
    plt.grid(True)

def plot_gender_bar_chart(gender_data):
    gender_count = Counter(gender_data)
    labels = list(gender_count.keys())
    values = list(gender_count.values())

    plt.subplot(1, 2, 2)
    plt.bar(labels, values, color='lightgreen', edgecolor='black')
    plt.title("Gender Distribution")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.grid(axis='y')

def main():
    print("This program will visualize age and gender distributions for multiple people.")
    age_data, gender_data = get_entity_data()

    plot_age_histogram(age_data)
    plot_gender_bar_chart(gender_data)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()


# In[ ]:





```
